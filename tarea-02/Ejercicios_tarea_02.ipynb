# Tarea 02


Crea un programa que imprima los números desde -30 hasta 50.

Para cada número, además de imprimirlo, el programa debe mostrar un texto según estas reglas:

si el número es menor que 0, mostrar "Muy frío"
si el número está entre 0 y 24, mostrar "Templado"
si el número es mayor o igual a 25, mostrar "Caluroso"

```
# clasificador de temperaturas
for temperatura in range(-30, 51): # recorre los numeros desde -30 hasta 50

    if temperatura < 0:
        print(temperatura, "Muy frío") # evalua si es menor a 0

    elif temperatura <= 24:
        print(temperatura, "Templado") # evalua si es hasta 24

    else:
        print(temperatura, "Caluroso") # para 25 o mas
```

# Problema 2. Números especiales del robot

Crea un programa que imprima los números del 1 al 100 que sean divisibles por 5.

Además, el programa debe cumplir estas condiciones:

si el número es divisible por 5, imprimir el número junto con el texto "Bip" si el número también es divisible por 10, imprimir el número junto con el texto "Bip Bip" Ejemplo de salida

```
# numeros especiales del robot
for numero in range(1, 101): # recorre del 1 al 100

    if numero % 10 == 0:
        print(numero, "Bip Bip") # si es divisible por 10

    elif numero % 5 == 0:
        print(numero, "Bip") # si es divisible por 5, pero no por 10
```

# Problema 3. Detector de puntaje gamer

Crea un programa que solicite al usuario dos números, los sume y luego muestre un mensaje según el resultado.

Reglas

si la suma es menor que 50, mostrar "Nivel principiante"
si la suma es mayor o igual a 50 y menor que 80, mostrar "Nivel intermedio"
si la suma es mayor o igual a 80, mostrar "Nivel pro"

```
# detector de puntaje gamer
puntaje1 = int(input("Ingresa primer puntaje: ")) # pide el primer numero
puntaje2 = int(input("Ingresa segundo puntaje: ")) # pide el segundo numero

suma = puntaje1 + puntaje2 # suma los puntajes

if suma < 50:
    print("Resultado: Nivel principiante") # evalua si es menor a 50

elif suma < 80:
    print("Resultado: Nivel intermedio") # evalua si es menor a 80

else:
    print("Resultado: Nivel pro") # para 80 o mas
```

#Problema 4. Calculador de perfil musical

Crea un programa que evalúe el perfil del usuario según su edad y sus gustos.

El usuario debe recibir los siguientes parámetros:

edad
le_gusta_musica
le_gusta_bailar
Y debe retornar uno de los siguientes mensajes:

"Eres una estrella de la fiesta" si la persona tiene 18 o más años, le gusta la música y le gusta bailar
"Te gusta la música, pero no tanto bailar" si le gusta la música pero no le gusta bailar
"Tienes tu propio estilo" en cualquier otro caso

```
# calculador de perfil musical
edad = int(input("Ingresa tu edad: ")) # pide la edad como numero
le_gusta_musica = input("¿Te gusta la música? (si/no): ").lower() # transforma a minuscula
le_gusta_bailar = input("¿Te gusta bailar? (si/no): ").lower() # transforma a minuscula

if edad >= 18 and le_gusta_musica == "si" and le_gusta_bailar == "si":
    print("Eres una estrella de la fiesta") # se cumplen las 3 condiciones

elif le_gusta_musica == "si" and le_gusta_bailar == "no":
    print("Te gusta la música, pero no tanto bailar") # musica si, baile no

else:
    print("Tienes tu propio estilo") # para cualquier otro caso
```
